```python
df = pd.read_csv('../../../international_armed_conflict/Data Sources/directed_dyadic_war.csv', encoding = 'latin-1')
df.rename({'statea': 'c_code_a', 'stateb': 'c_code_b'}, axis = 1, inplace = True)

c_code_df = pd.read_pickle('../../../international_armed_conflict/Pickles/c_code_df.pkl')
c_code_df.rename({'c_code': 'c_code_a', 'state_name': 'state_name_a'}, axis = 1, inplace = True)

war_df = deepcopy(df[(df['warnum']==106) & (df['durindx']==1)])
war_df = deepcopy(pd.merge(c_code_df, war_df, on = 'c_code_a'))

c_code_df.rename({'c_code_a': 'c_code_b', 'state_name_a': 'state_name_b'}, axis = 1, inplace = True)

war_df = deepcopy(pd.merge(c_code_df, war_df, on = 'c_code_b'))
war_union_df = deepcopy(war_df)
war_union_df.rename({'c_code_a': 'c_code_new_a', 'c_code_b': 'c_code_new_b', 'state_name_a': 'state_name_new_a', 'state_name_b': 'state_name_new_b', 'batdtha': 'batdtha_new', 'batdthb': 'batdthb_new'}, axis = 1, inplace = True)
war_union_df.rename({'c_code_new_a': 'c_code_b', 'c_code_new_b': 'c_code_a', 'state_name_new_a': 'state_name_b', 'state_name_new_b': 'state_name_a', 'batdtha_new': 'batdthb', 'batdthb_new': 'batdtha'}, axis = 1, inplace = True)
```

```python
war_df = deepcopy(pd.concat([war_df, war_union_df]))
war_df.sort_values(by = 'year', ascending = True, inplace = True)

duplicate_list = ['c_code_a']

war_df.drop_duplicates(subset = duplicate_list, keep = 'first', inplace = True)
war_df = deepcopy(war_df[['c_code_a', 'state_name_a', 'c_code_b', 'state_name_b', 'batdtha', 'batdthb', 'year']].reset_index(drop=True))
```

```python
network_nodes = list(war_df['state_name_a'])

graph_file = open("network_of_wars.json", 'w+')

json_dic = {}
json_dic['nodes'] = {}
json_dic['links'] = {}
start_line = '{\n  "nodes": [\n'
middle_line = '  ],\n  "links": [\n'
end_line = '\n  ]\n}'

graph_file.write(start_line)
for i, node in enumerate(war_df['state_name_a']):
    if node == network_nodes[-1]:
        add_line = '    {"country": "' + node + '", "size": ' + str(war_df.loc[i, 'batdtha']) + '}\n'
        graph_file.write(add_line)
    else:
        add_line = '    {"country": "' + node + '", "size": ' + str(war_df.loc[i, 'batdtha']) + '},\n'
        graph_file.write(add_line)        

graph_file.write(middle_line)

for i, node_1 in enumerate(war_df['state_name_a']):
    if i+1 != len(network_nodes):
        add_line = '    {"source": ' + str(network_nodes.index(node_1)) + ', "target": ' + str(network_nodes.index(war_df.loc[i, 'state_name_b'])) + ', "bond": ' + str(1) + '},\n'
#         add_line = '    {"source": "' + str(node_1) + '", "target": "' + str(war_df.loc[i, 'state_name_b']) + '", "bond": ' + str(1) + '},\n'
        graph_file.write(add_line)
    else:
        add_line = '    {"source": ' + str(network_nodes.index(node_1)) + ', "target": ' + str(network_nodes.index(war_df.loc[i, 'state_name_b'])) + ', "bond": ' + str(1) + '}\n'
#         add_line = '    {"source": "' + str(node_1) + '", "target": "' + str(war_df.loc[i, 'state_name_b']) + '", "bond": ' + str(1) + '}\n'
        graph_file.write(add_line)

graph_file.write(end_line)
```
